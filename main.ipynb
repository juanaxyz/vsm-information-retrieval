# Fungsi Hitung TF - IDF, Weight

```python
def hitung_tf(doc):
    # TF mentah (jumlah kemunculan term dalam 1 dokumen)
    freq = Counter(doc)
    return {t: freq.get(t, 0) for t in term}

def hitung_df_per_term(t, docs_list):
    # DF = jumlah dokumen yang mengandung term t (bukan jumlah frekuensi)
    return sum(1 for d in docs_list if t in d)
    
```

```python
# DataFrame dokumen asli vs hasil preprocessing
df_docs = pd.DataFrame({
    "dokumen_asli": docs,
    "dokumen_preprocessed": [" ".join(doc) for doc in preprocessed_docs]
})

# DataFrame TF dibalik: kolom = dokumen, baris = term
df_tf = pd.DataFrame(
    [hitung_tf(doc) for doc in preprocessed_docs],
    index=[f"D{i+1}" for i in range(N)]
).T

# Tambah kolom TF untuk query
df_tf["Q"] = pd.Series(hitung_tf(query_preprocessed))



df=df_tf.copy()
df["DF"]=[hitung_df_per_term(t, preprocessed_docs) for t in df.index]


tf_idf = df.copy()
tf_idf["IDF (log N/df)"] = tf_idf["DF"].apply(lambda df: math.log(N / df) + 1 if df > 0 else 0)


def hitung_weight(tf, idf):
    return tf * idf

# Hitung bobot (W) untuk semua dokumen D1..D7
columns=[f"D{i+1}" for i in range(N)]
for col in columns:  # columns = ['D1', ..., 'D7']
    tf_idf[f"W_{col}"] = tf_idf.apply(
        lambda row: hitung_weight(row[col], row["IDF (log N/df)"]),
        axis=1
    )

# Jika ingin juga bobot query, aktifkan baris ini:
tf_idf["W_Q"] = tf_idf.apply(lambda row: hitung_weight(row["Q"], row["IDF (log N/df)"]), axis=1)
```

## Keterangan Komponen Cosine Similarity

- **Sum(Q × D)**: jumlah hasil perkalian bobot query dan bobot dokumen pada setiap term.
- **|Q|**: panjang (norma) vektor query.
- **|D|**: panjang (norma) vektor dokumen.

\[
\cos(Q, D)=\frac{\sum (Q_i \times D_i)}{|Q| \times |D|}
\]

# Menghitung Cosine Similarity untuk SVM

```python
# Tabel detail seperti perhitungan manual: Q^2, D_i^2, dan Q*D_i
detail = pd.DataFrame(index=tf_idf.index)
detail["Q^2"] = tf_idf["W_Q"] ** 2

for d in columns:
    detail[f"{d}^2"] = tf_idf[f"W_{d}"] ** 2
    detail[f"Q*{d}"] = tf_idf["W_Q"] * tf_idf[f"W_{d}"]

print("Tabel detail cosine (per term):")
display(detail.round(6))

# Rekap sum dan cosine similarity per dokumen
sum_q2 = float(detail["Q^2"].sum())
sqrt_q = math.sqrt(sum_q2)

hasil = []
for d in columns:
    sum_d2 = float(detail[f"{d}^2"].sum())
    sqrt_d = math.sqrt(sum_d2)
    sum_qd = float(detail[f"Q*{d}"].sum())
    cosine = sum_qd / (sqrt_q * sqrt_d) if sqrt_q > 0 and sqrt_d > 0 else 0.0
    hasil.append({
        "dokumen": d,
        "Sum(Q*D)": sum_qd,
        "|Q|": sqrt_q,
        "|D|": sqrt_d,
        "cosine_similarity": cosine
    })

df_cosine = pd.DataFrame(hasil).sort_values("cosine_similarity", ascending=False)

print("Ranking cosine similarity:")
display(df_cosine.round(6))
```
